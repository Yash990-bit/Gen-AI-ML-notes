```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.xlabel('CGPA')
plt.ylabel('Placement_exam_marks')
plt.show()
```

```python
X = df.iloc[:,0:1]
y=df.iloc[:,-1]
y
```

# Linear Regression Model Fromula

 Y = 0+1X+

 Y = dependent(target/output) ,
 X = independent(input/feature) ,
 0 = Intercept(constant) ,
 1 = Slope(coffiecent) ,
  = Error term(noise) ,

# Why we use Linear Regression & When to Use It ?

 Predict a number (continuous value) based on input features.


*   Output is Continuous (Numbers)
*   Relationship is Linear (Straight Line Trend)
*   Simple & Interpretable Model Needed





```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
X
```

```python
y
```

```python
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.plot(X,lr.predict(X),color='red')
plt.xlabel('CGPA')
plt.ylabel('Placement_exam_marks')
plt.show()
```

```python
m = lr.coef_ #Slope
m
```

```python
b = lr.intercept_ #Intercept
b
```

```python
# y = mx+b

m * 7.19 + b
```
