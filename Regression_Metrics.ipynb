# MAE (Mean Absolute Error)

```python
# Formual

MAE   =    summation|Yi - ^Y|/n

Yi = Actual Value
^
Y = Predicted value

n = Number of data points

MSE cares more about big errors than small errors.
```

```python
#Why Do We Square the Error?

Remove negative values (no canceling)

Punish large mistakes more
```

```python
#Where MSE Is Used

Linear Regression

House price prediction

Salary prediction

Sales forecasting

Any continuous number prediction
```

# MSE (Mean Squared Error)


```python
#Formula

MSE   =    summation(Yi - ^Y)^2/n

Yi = Actual Value
^
Y = Predicted value

n = Number of Samples

MSE = Average of squared mistakes
Bigger mistake → Much bigger punishment
```

```python
#Why Square the Error?

Removes negative signs

Penalizes big mistakes more
```

```python
#Where MSE is Used?

Linear Regression

House price prediction

Salary prediction

Sales forecasting

Any continuous number prediction
```

# RMSE (Root Mean Squared Error)

```python
# Formula


RMSE   =    √summation(Yi - ^Y)^2/n

Yi = Actual Value
^
Y = Predicted value

n = Number of data points

RMSE = Average prediction error in real units.
RMSE = square root of MSE = real-world error
```

```python
#Why We Use RMSE Instead of Only MSE

MSE → squared units (hard to understand)

RMSE → original units (easy to interpret)
```

# R2 Score

```python
# Formula

R2 = 1-SSr/SSm

SSr = Sum of squared error in regression line
SSm = Sum of squared error in Mean Line

when you compare worst of mean line and best to regression line = R² score”
If SSr is greater then SSm
then R2 Score is (-ve)

The mistake in R2 score is when we add more input in the data then r2 socre is increasing that the reason it fails sometime
```

# Code

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

```python
# Residual = Actual_value − Predicted value



residual = y_test-y_pred
residual
```
