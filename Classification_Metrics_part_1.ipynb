  accuracy = no.of correct prediction /total prediction

  how many predictions the model got correct out of total predictions.

# Where it is used for
Classification models (Logistic Regression, Decision Tree, etc.)

Spam detection

Fraud detection

Disease prediction

# Not Used for

regression (like R², MSE are used there).

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SUV_data.csv')
```

```python
df.head()
```

```python
X = df.drop("Salary", axis=1)  # features
y = df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

*  Not Suitable for Imbalanced Real-World Problems

*  Not Used for Regression Problems - only for classification

* For regression (like price prediction, sales prediction),   

   R² Score

   MAE

  RMSE
