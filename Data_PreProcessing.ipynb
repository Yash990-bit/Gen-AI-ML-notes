# Trimming


```python
#Trimming

trim = df[(df['Glucose']>8.80) | (df['Glucose']<5.11)]
trim
```

```python
# Approach 2

#Calculating the Z-Score (X-mu/sigma)

df['cg'] = (df['Glucose']-df['Glucose'].mean()/df['Glucose'].std())
df
```

# Capping

```python
#Capping

upper = df['Glucose'].mean()+3*df['Glucose'].std()
lower = df['Glucose'].mean()-3*df['Glucose'].std()

```

```python
df['Glucose'] = np.where(
    df['Glucose']>upper,
    upper,
    np.where(
        df['Glucose']<lower,
        lower,
        df['Glucose'] )
)
```

# Outlier detection using IQR


```python
# Finding the IQR

Q1 = df['Glucose'].quantile(0.25)
Q3 = df['Glucose'].quantile(0.75)
```

```python
# Q1
# Explanation for OrdinalEncoder (oe) error:
# The ValueError occurred because OrdinalEncoder was initialized with string categories
# (e.g., ['Poor','Average','Good']) but then tried to fit on numerical columns (Pregnancies, Glucose) in X_train.
# To resolve this, ensure that the columns you are applying OrdinalEncoder to are indeed categorical (e.g., 'Gender', 'Education Level')
# and that the categories provided to the encoder match the unique values present in those columns.
Q1
```

```python
Q3
```

```python
IQR = Q3-Q1 #(Calculate IQR)

IQR
```

```python
lower = Q1-1.5*IQR
upper = Q3+1.5*IQR

lower
upper
```

```python
#Find Outliers

clean = df[(df['Glucose']>upper) | (df['Glucose']<lower)]
clean
```
