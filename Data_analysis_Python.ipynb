# Groupby

Group similar values together and then perform some calculation on them

If you have student marks data and you want:

Average marks per class

Total salary per department

Total sales per month

# We use Groupby()

```python
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Department": ["IT", "HR", "IT", "HR", "IT"],
    "Salary": [50000, 40000, 60000, 45000, 70000]
}
df = pd.DataFrame(data)
print(df)
```

# Total Salary Per Department

```python
df.groupby('Department')['Salary'].sum()
```

# Average Salary Per Department

```python
df.groupby('Department')['Salary'].mean()
```

# For function

# Lambda in GroupBy

Because sometime we built

.sum()

.mean()

.max()

```python
# Salary Range Per Department

# Range = max-min

df.groupby("Department")["Salary"].agg(lambda x: x.max() - x.min())
```

# groupby.filter


```python
df.groupby("Department").filter(lambda x: x["Salary"].mean() > 50000)
```

# Sort_index

```python
df.groupby("Department")["Salary"].sum().sort_index(ascending=True)

# If ascending=False → reverse order.
```

```python
df.groupby("Department")["Salary"].sum().sort_values(ascending=False)

# If ascending=True → smallest revenue first.
```

 sort_index() → sorts by group names (index)

sort_values() → sorts by aggregated values

ascending=True → small → large

ascending=False → large → small

# Indexing array

# 1️⃣ Indexing starts at 0

First element → index 0

Second element → index 1

Third element → index 2

# Example

arr = [10, 20, 30, 40]

arr[0] → 10

arr[2] → 30

# 2️⃣ Slicing: arr[start:end]

start → inclusive (included)

end → exclusive (stop before this index)

# Exmaple

arr[:3]   # first three elements → [10, 20, 30]

arr[1:4]  # index 1,2,3 → [20, 30, 40]

# 3️⃣ Negative index & steps

Negative index counts from end:

-1 → last element

-2 → second last element

Step / skip elements: arr[start:end:step]

# Exmaple

arr[-2:]   # last 2 elements → [30, 40]

arr[::2]   # every 2nd element → [10, 30]

# Bonus point remember

“Start included, End excluded” → always stops before end index.
If start > end → empty array

# drop_duplicate

Removes duplicate rows from a DataFrame

Keeps only the first occurrence (by default)

Can remove duplicates based on all columns or specific columns

# Subset

Which column(s) to check for duplicates. Default is None → check all columns

# keep

Which duplicate to keep: 'first', 'last', or False (drop all duplicates)

# inplace

If True → changes the original DataFrame; if False → returns new DataFrame




```python
# Syntax

df.drop_duplicates(subset=None, keep='first', inplace=False)
```
