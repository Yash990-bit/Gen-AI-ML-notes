```python
import numpy as np
import matplotlib.pyplot as plt #Graph
def sigmoid(z):
  return 1/1+(np.exp(-z))

#Generate Value
z = np.linspace(-10, 10, 100)
y = sigmoid(z)

plt.title("Sigmoid Function")
plt.plot(z, y)
plt.ylabel("sigmoid(z)")
plt.show()
```

# Loss Function

Why Do We Need a Loss Function in Logistic Regression?

Because the model outputs a probability (0 to 1), not a direct number.

Formula (Entropy)

Loss= −[ylog(p)+(1−y)log(1−p)]

Where:

y = Actual label (0 or 1)

p = Predicted probability
